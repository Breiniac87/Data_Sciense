```python
import numpy as np
from PIL import Image,ImageOps
import matplotlib.pyplot as plt
from wordcloud import WordCloud,STOPWORDS,ImageColorGenerator
from scipy.ndimage import gaussian_gradient_magnitude
import pandas as pd
```

```python
df = pd.read_excel('nap.xlsx')
```

```python
def multiply_word(a, b):
    result = (a + ' ') * b

    return result.strip()
```

```python
for a, b in df[['word', 'частотность']].values:
  with open('result.txt','a') as file:
    file.write('\n ')
    file.write(multiply_word(a, b))
```

```python
file=open('result.txt','r')
text=file.read()
```

```python
text
```

```python
canvas_width=1920
canvas_height=1080
stopwords = set(STOPWORDS)
```

```python
wordcloud = WordCloud(repeat=False, max_words=200,stopwords=stopwords, width=canvas_width,height=canvas_height, collocations=False, relative_scaling=0.1).generate(text)
plt.imshow(wordcloud, interpolation='[]')
```

```python
wordcloud.to_file('napitki.png')
```
