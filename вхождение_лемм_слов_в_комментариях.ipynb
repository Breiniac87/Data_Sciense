```python
new_df = pd.DataFrame(columns= ['ID', 'word', 'text'])
for ID, report, proposal in df[['ID', 'report', 'lemmatize_text']].values:
  for word in proposal.split(' '):
    if word in df_unique_w.word.values:
      # print(ID,word, report)
  # break
      new_df = new_df.append({'ID': ID, 'word': word, 'text': report}, ignore_index=True)
```

```python
new_df
```

```python
result = pd.merge(new_df, df_unique_w, on='word', how='left')
```
